```python
from nltk.corpus import stopwords
```

```python
def preprocessing(txt):
    txt = txt.lower()                   # all characters in lower
    txt = re.sub('[^a-zA-Z]',' ',txt)   # all characters in string format
    txt = re.sub('http\S+\s*',' ',txt)  #link removal
    txt = re.sub('RT|cc', ' ', txt) 
    txt = re.sub('#\S+',' ',txt)        # #tag removal
    txt = re.sub('\s+',' ',txt)         # extra whitespace removal
    txt = re.sub('@\S+',' ',txt)        # @mentions removal
    txt = nltk.tokenize.word_tokenize(txt) # tokenizing words
    txt = [i for i in txt if not i in stopwords.words('english')]   # removal of stopwords
    return ' '.join(txt)                  # joining texts
     
```

```python
df['Resumes'] =df['Resumes'].apply(lambda i :preprocessing(i))
```

```python
import string
from nltk.corpus import stopwords
```

```python
# collection of most common words, after removing stopwords and punctuation.

Setofstopwords =set(stopwords.words('english')+['``',"''"])  #set of stopwords

totalwords =[]

sentences = df['Resumes'].values
cleanstentences =""             # empty string

for records in sentences:
    cleanstentences+=records     # concate all resumes in empty string
    requiredwords = nltk.word_tokenize(records)  #tokenize words 
    for word in requiredwords:
        if word not in Setofstopwords and word not in string.punctuation:
            totalwords.append(word)
freqwordlist = nltk.FreqDist(totalwords)    #created a frequency distribution of words
mostfrquent = freqwordlist.most_common(100)   # 100 most common word with their frequencies
print(mostfrquent)
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()

lem_words =[]  # empty list to stored words with same meaning(lemmatize words)

for word in freqwordlist:
    word = lemma.lemmatize(word)
    lem_words.append(word)
```

```python
def wordfreq(df):
    count = df['Resumes'].str.split(expand=True).stack().value_counts().reset_index()
    count.columns =['Word','Frequency']
    
    return count.head(10)
```

```python
fig = plt.figure(figsize=(30,60))

for i ,category in enumerate(categories):
    wf = wordfreq(df_categories[i]) 
    
    fig.add_subplot(4,4,i+1).set_title(category)
    plt.bar(wf['Word'],wf['Frequency'])
    plt.ylim(0,500)
plt.tight_layout()
plt.show()
plt.close()
```

# MODEL Building

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
```

```python
x = df['Resumes'].values
y = df['Final_Category'].values

tf = TfidfVectorizer(sublinear_tf=True,stop_words='english')
tf.fit(x)
tf = tf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(tf,y,test_size=0.20,random_state=42,stratify=y)
```

# Importing ML Algorithms 

# Using Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
```

```python
rf = RandomForestClassifier()
rf.fit(xtrain,ytrain)
pred_rf = rf.predict(xtest)

print('Accuracy for RF in training set',rf.score(xtrain,ytrain))
print('Accuracy for RF in testing set',rf.score(xtest,ytest))

print(classification_report(ytest,pred_rf))

```

```python
train_1 = rf.score(xtrain,ytrain)
test_1 = rf.score(xtest,ytest)
accuracy_1 =accuracy_score(ytest,pred_rf) 
recall_1 = recall_score(ytest,pred_rf,average='macro')
precision_1 = precision_score(ytest,pred_rf,average='macro')
f1_1  =f1_score(ytest,pred_rf,average='macro')
```

```python
print('Training Accuracy',train_1)
print('Testing Accuracy',test_1)
print('Accuracy score =',accuracy_1)
print('Recall Score =',recall_1)
print('Precision Score=',precision_1)
print('F1 Score =',f1_1)
print('confusion_matrix =\n\n',confusion_matrix(ytest,pred_rf))
```
